# Turbidity Plot

```python
tamanoletra = 10
estiloletra = "Arial"

fig, (ax1) = plt.subplots(1,1,figsize=(15,3),dpi=300)

ax1.scatter(data_BotCr_below.date,data_BotCr_below.turbidity,color='r',label='Below')
ax1.scatter(data_BotCr_above.date,data_BotCr_above.turbidity,color='y',label='Above')
ax1.scatter(data_BotCr_along.date,data_BotCr_along.turbidity,color='g',label='Along')

# ax1.scatter(data_BlackWat_below.date,data_BlackWat_below.turbidity,color='r',label='Below')
ax1.set_ylabel(r'Turbidity',color='k',fontsize=tamanoletra,fontname=estiloletra)
# ax1.tick_params(axis='y',labelcolor='k',labelsize=tamanoletra)
ax1.set_xlabel(r'Date',color='k',fontsize=tamanoletra,fontname=estiloletra)
# ax1.tick_params(axis='x',labelcolor='k',labelsize=tamanoletra)
plt.legend(loc="best",fontsize=tamanoletra)
```
